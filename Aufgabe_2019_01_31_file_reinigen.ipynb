```python
import numpy as np

def Datei_einlesen(Pfad):
    gereinigte_Zeilen = []
    file = open(Pfad)
    for line in file:
        string = line

        # lösche Zeilenumbruch: "... /n"
        string = string[:-1]

        # lösche Komentare: "... | % 53"
        if "|" in string:
            string = string[:string.find("|")]

        # lösche Komentare: "... %barkeysig"
        if "%" in string:
            string = string[:string.find("%")]

        # lösche Komentare: "... \repeat volta 2"
        if "\\repeat" in string:
            string = string[:string.find("\\repeat")]

        # lösche Sonderzeichen: "} { . ' , "
        for zeichen in ["}", "{", ".", "'", ",", "~", "\""]:
            string = string.replace(zeichen, "")    

        # lösche Zahlen: "f8( \p a) f( a) f( a) "
        for zeichen in ["2", "4", "8"]:
            string = string.replace(zeichen, "")    

        # lösche Leerzeilen: "   "
        if len(string.strip()) == 0:
            continue

        # lösche Befehlzeilen: "\time 6/8"
        if string.strip()[0]=="\\" :#or "=" in Zeile:
                continue

        # lösche Befehle in Noten: "f8( \p a) f( a) f( a"
        while(string.find("\\")!=-1):
            n = 0
            while(string[string.find("\\")+n]!=" " and len(string)>string.find("\\")+n+1):
                n += 1
            if len(string)>string.find("\\")+n+2:
                string = string[:string.find("\\")]+string[string.find("\\")+n+1:]
            else:
                string = string[:string.find("\\")]+string[string.find("\\")+n:]

       # lösche Anfangs- und Endfreizeichen : "  ... "
        string = string.strip()
        gereinigte_Zeilen.append(string)
    file.close()
    return gereinigte_Zeilen


Zeilen = Datei_einlesen("Metamorphoses_Three_02_easy.ly.sec")
np.savetxt("Metamorphoses_Three_02_easy_clean.out", Zeilen, fmt="%s")
```
